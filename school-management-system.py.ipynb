```python
#Defining remove_student function

def remove_student():
    class_studying = input("class name:")
    roll = input("Roll No:")
    
    data = (class_studying,roll)
    
    sql = 'DELETE FROM student WHERE class_studying=%s and roll=%s'
    c = con.cursor()
    c.execute(sql,data)
    con.commit()
    print("Data Updated")
    print("--------------------------------------------------------------------------------------------------")

    main()
```
